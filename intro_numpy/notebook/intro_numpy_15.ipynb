# プロットへの注釈の追加

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([[3, 0.53, 4.0753, 0.5],
              [10, 1.78, 6.6678, 2.1328],
              [19, 0.86, 1.5177, 3.6852],
              [20, 1.6, 3.6375, 8.5389],
              [21, 3, 4.7243, 10.157],
              [23, 6.11, 9.0698, 2.8739],
              [38, 2.54, 5.3002, 4.4508]])
sample = np.copy(data[:,0])
density = np.copy(data[:,1])
v1 = np.copy(data[:,2])
v2 = np.copy(data[:,3])
mass1 = density*v1
mass2 = density*v2
````

matplotlib.pyplot の title などのプロット注釈関数を使用して、プロットにラベルを追加できます。これらの関数への入力は文字列です。
```
plt.plot("Plot Title")
```
タイトル "Sample Mass" を追加してみましょう。



````python
plt.plot(sample,mass1,'ks')
plt.plot(sample,mass2,'r*')
````

関数 ylabel を使用して、ラベル 'Mass (g)' を追加しましょう、

関数 legend を使用すると、プロットに凡例を追加できます。

```
legend('a','b','c')
```
ラベル 'Exp A', 'Exp B' をこの順序で含む凡例を追加しましょう。

string を変数と「連結」することにより、定数の値をプロット注釈で利用できます。

````python
plt.bar(np.arange(len(data[2,:])),data[2,:])
plt.title('Sample ' + str(int(sample[2])) + ' Data')
````

単一のベクトルを単独でプロットするとき、matplotlib.pyplot では、ベクトルの値を y軸のデータとして使用し、x 軸のデータを 0 から n-1 (ベクトル内の要素数は n)までの範囲に設定します。

以下のコマンドを使用してベクトル v1 をプロットしましょう。
```
plt.plot(v1)
```

関数 plot は、キーワード引数を追加できます。
```
plt.plot(y, linewidth=5)
```
上記のコマンドは太い線をプロットします。使用可能なプロパティについては、np.info(plt.plot) などを参照してください。

v1 を、ライン幅 3 でプロットしましょう。



関数 plot では、プロットされる引数とライン指定引数の後に、キーワード引数を指定できます。

v1('y軸')を sample('x軸')に対して、赤(r)の円形(o)マーカ付き、ライン幅 4 でプロットしましょう。

関数 plot は折れ線グラフを作成します。matplotlib.pyplot には他にも多数のプロット関数があります。詳細なリストについては、[matplotlib reference](https://matplotlib.org/stable/api/pyplot_summary.html) などを参照してください。

プロットごとにキーワード引数は異なります。関数 hist を使用して、density のヒストグラムを作成してみてください。color を黄色('y')に設定しましょう。
