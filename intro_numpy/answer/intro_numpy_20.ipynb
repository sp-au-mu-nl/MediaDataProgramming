# 判定分岐

````python
import numpy as np
import matplotlib.pyplot as plt
import random

doPlot = random.randint(0,1)
data = np.array([[0.53, 4.0753, np.nan],
              [1.78, 6.6678, 2.1328],
              [0.86, 1.5177, 3.6852],
              [1.6, 3.6375, 8.5389],
              [3, 4.7243, 10.157],
              [6.11, 9.0698, 2.8739],
              [2.54, 5.3002, 4.4508]])
element = ['Lithium',
'Argon',
'Potassium',
'Calcium',
'Scandium',
'Vanadium',
'Strontium',]
density = data[:,0]
````

if ブロックの本体は、条件が True の場合にのみ実行されます。

(これ以前のものはコードの書き方が不適切)
```
import random
x = random.random()
if x > 0.5: # 条件
     y = 3 # 本体
```
doPlot が 1 の場合にのみ下記のプロット用コードが実行されるように、スクリプトを変更します。

````python
if doPlot:
  plt.plot(density)
  plt.title('Sample Densities')
  plt.xticks(np.arange(len(element)),element)
  plt.ylabel('Density ($g/cm^3$)')
````

条件が満たされない時は、他のコードを実行する場合があります。これには、else キーワードを使用します。
```
import random
x = random.random()
if x > 0.5:
    y = 3
else:
    y = 4
```

スクリプトを変更し、if 条件が満たされない場合に以下のコード行が実行されるようにします。
```
for e, d in zip(element, density):
    print("The density of " + e + " is " + str(d))
```

````python
if doPlot:
  plt.plot(density)
  plt.title('Sample Densities')
  plt.xticks(np.arange(len(element)),element)
  plt.ylabel('Density ($g/cm^3$)')
else:
  for e, d in zip(element, density):
    print("The density of " + e + " is " + str(d))
````


elseif キーワードを if の後に使用すると、条件を追加できます。elseif ブロックは複数含めることができます。
